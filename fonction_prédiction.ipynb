```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv('data_pour_predictions.csv')
df = df.drop(['Unnamed: 0', 'ACCESSION'],1)
df.head(8)


```

```python
#### preprocessing

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df[['ion_precursor_mz','num_peak','peak_mz','peak_rel_int']] = \
scaler.fit_transform(df[['ion_precursor_mz','num_peak','peak_mz','peak_rel_int']])
```

```python
### entrainement des donn√©es

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split


train_df, test_df = train_test_split(df, test_size = .1)

drop = ['class1', 'class2', 'class3']

X_train = train_df.drop(drop, axis=1)
Y_train1 = train_df["class1"]
Y_train2 = train_df["class2"]
Y_train3 = train_df["class3"]

X_test  = test_df.drop(drop, axis=1).copy()
Y_test1 = test_df["class1"]
Y_test2 = test_df["class2"]
Y_test3 = test_df["class3"]   

#class1
random_forest_1 = RandomForestClassifier(criterion = "entropy", 
                                       min_samples_leaf = 1, #min number of data points allowed in a leaf node
                                       min_samples_split = 2, #min number of data points placed in a node before the node is split  
                                       n_estimators=100, #number of trees in the foreset
                                       max_features='auto', #max number of features considered for splitting a node
                                       oob_score=True, 
                                       n_jobs=-1)

random_forest_1.fit(X_train, Y_train1)
Y_prediction1 = random_forest_1.predict(X_test)


#class2
random_forest_2 = RandomForestClassifier(criterion = "entropy", 
                                       min_samples_leaf = 1, #min number of data points allowed in a leaf node
                                       min_samples_split = 2, #min number of data points placed in a node before the node is split  
                                       n_estimators=100, #number of trees in the foreset
                                       max_features='auto', #max number of features considered for splitting a node
                                       oob_score=True, 
                                       n_jobs=-1)

random_forest_2.fit(X_train, Y_train2)
Y_prediction2 = random_forest_2.predict(X_test)

#class3
random_forest_3 = RandomForestClassifier(criterion = "entropy", 
                                       min_samples_leaf = 1, #min number of data points allowed in a leaf node
                                       min_samples_split = 2, #min number of data points placed in a node before the node is split  
                                       n_estimators=100, #number of trees in the foreset
                                       max_features='auto', #max number of features considered for splitting a node
                                       oob_score=True, 
                                       n_jobs=-1)

random_forest_3.fit(X_train, Y_train3)
Y_prediction3 = random_forest_3.predict(X_test)
```

```python
def __prediction_classe_3__(ion_precursor_mz, num_peak, peak_mz, peak_rel_int) :
    
    dict_pred = {'ion_precursor_mz' : ion_precursor_mz,
                 'num_peak' : num_peak,
                 'peak_mz' : peak_mz,
                 'peak_rel_int' : peak_rel_int}
    
    df_a_predire = pd.DataFrame({key:pd.Series(value) for key, value in dict_pred.items()})
    df_a_predire['ion_precursor_mz'] = df_a_predire['ion_precursor_mz'].fillna((df_a_predire['ion_precursor_mz'][0]))
    df_a_predire['num_peak'] = df_a_predire['num_peak'].fillna((df_a_predire['num_peak'][0]))
    
    df_a_predire = scaler.transform(df_a_predire)
    
    predictions = random_forest_3.predict(df_a_predire)
    return  predictions

__prediction_classe_3__(ion_precursor_mz =[316.0], num_peak=[2], peak_mz=[299.0, 180.0, 163.0, 145.0, 137.0, 119.0, 91.0, 85.0],\
               peak_rel_int=[3, 10, 100, 60, 15, 20, 10, 35])

```
