```python
df2 = df.copy()
```

```python
print("nombre de valeur null", df2.isnull().sum().sum())
df2.dropna(how='any', axis=1, inplace=True)
print("nombre de valeur null apres dropna", df2.isnull().sum().sum())
```

```python
del df2['Unnamed: 0'] 
del df2['Unnamed: 0.1'] 
df2.head(2)
```

```python
df2.info()
```

```python
unique_users = df2['ACCESSION'].unique()
target_n = int(.8 * len(df2))
shuffled_users = np.random.permutation(unique_users)

user_count = df2['ACCESSION'].value_counts()

mapping = user_count.reindex(shuffled_users).cumsum() <= target_n
mask = df2['ACCESSION'].map(mapping)

df_train = df2[mask]
df_test = df2[~mask]
```
